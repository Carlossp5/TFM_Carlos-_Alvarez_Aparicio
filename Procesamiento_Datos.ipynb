# **0. Cargar librerias**


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
pd.set_option('display.max_columns',None)
```

# **1. Cargar Datasets**

```python
GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28 = pd.read_csv('/content/drive/MyDrive/Master Analisis de datos futbol/Master/Tablas_FBREF/GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28.csv', sep=',', index_col=False)
print(GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28.shape)
GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28.head()
```

# **2. Procesamiento de los datos**

```python
# Hay 3 tipos de variables: absolutas, por partido y porcentuales. Las identifico porque las voy a tratar para escalarlas todas entre 0 y 100
columnas = GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28.columns.to_list()
variables_porcentuales = ['Posesion_media_a_favor ','Save%','CS%_a_favor','Penaltis_Save%','PSxG/SoT','Porc_Pases_Largos_Completados_Portero','Porc_Pases_Largos_Portero','Porc_Saques_Puerta_Largo','Porc_Centros_Detenidos_PT','Goles_por_disparo_a_puerta','xG_sin_penaltis_por_tiro','Porc_Pases_Completados_a_favor', 'Porc_Pases_Cortos_Completados_a_favor','Porc_Pases_Medios_Completados_a_favor','Porc_Pases_Largos_Completados_a_favor',
                          'Porc_Tackle_Ganado_a_favor','Posesion_a_favor','Porc_Regates_Completados_a_favor','Porcentaje_Tackles_Regate_en_contra','Porc_Duelos_Aereos_ganados_equipo_propio',
                          'Posesion_media_en_contra','Porc_SoT_Detenidos_en_contra','CS%__en_contra','Porc_PK_Detenidos_en_contra','PSxG/SoT_a_favor','Porc_Pases_Largos_Completados_Portero_rival','Porc_Pases_Largos_Portero_rival','Porc_Saques_Puerta_Largo_rival','Porc_Centros_Detenidos_PT_rival','Goles_por_disparo_en_contra.1','xG_en_contra_sin_penaltis_por_tiro','Porc_Pases_Completados_en_contra', 'Porc_Pases_Cortos_Completados_en_contra','Porc_Pases_Medios_Completados_en_contra','Porc_Pases_Largos_Completados_en_contra',
                          'Porc_Tackle_Ganado_en_contra','Posesion_en_contra','Porc_Regates_Completados_en_contra','Porcentaje_Tackles_Regate_a_favor','Porc_Duelos_Aereos_ganados_equipos_rivales']
variables_por_partido = ['xGD/90','Gls/90_a_favor','Ast/90_a_favor','G-PK/90_a_favor','xG/90_a_favor','xAst/90_a_favor','xG+xAG/90_a_favor','npxG/90_a_favor','npxG+xAst/90_a_favor','GA/90','PSxG_por_partido_en_contra','Acciones_Fuera_Area_Por_Partido', 'Tiros_a_favor_por_partido','Tiros_a_puerta_a_favor_por_partido','Acciones_Creadas_Gol_por_partido_a_favor',
                        'Gls/90__en_contra','Ast/90__en_contra','G-PK/90__en_contra','xG/90__en_contra','xAst/90__en_contra','xG+xAG/90__en_contra','npxG/90__en_contra','npxG+xAst/90__en_contra','PSxG_por_partido_a_favor','Acciones_Fuera_Area_PT_rival_Por_Partido','Tiros_en_contra_por_partido','Tiros_a_puerta_en_contra_por_partido','Acciones_Creadas_Gol_por_partido_en_contra']
variables_absolutas = [columna for columna in columnas if columna not in variables_porcentuales and columna not in variables_por_partido]
variables_absolutas = variables_absolutas[4:]
```

```python
# Pongo las variables absolutas como variables por partido
GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_var_abs_std = pd.DataFrame()
for col in variables_absolutas:
  GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_var_abs_std[col] = GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28[col]/GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28['MP']
```

```python
# Uno las variables por partudo y las reordeno
GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_var_abs_std = pd.concat([GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_var_abs_std,GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28[variables_por_partido],GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28[variables_porcentuales]], axis=1)
```

```python
# Corrijo las variables porcentuales que tienen valores entre 0 y 1 conviertiendolas a entre 0 y 100
variables_porcentuales_corr = ['PSxG/SoT','Goles_por_disparo_a_puerta','xG_sin_penaltis_por_tiro','PSxG/SoT_a_favor','Goles_por_disparo_en_contra.1','xG_en_contra_sin_penaltis_por_tiro']
GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_var_abs_std[variables_porcentuales_corr] = 100*GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_var_abs_std[variables_porcentuales_corr]
```

```python
# Reescalo todas las variables para que esten entre 0 y 100
scaler = MinMaxScaler(feature_range=(0, 100))

# Escalar los datos
GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_scaled = pd.DataFrame(scaler.fit_transform(GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_var_abs_std), columns=GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_var_abs_std.columns)
```

```python
# Uno las variables ya escaladas con los identificadores
GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_scaled = pd.concat([GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28.iloc[:,0:4],GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_scaled], axis=1)
```

```python
# Chequeamos que todas las variables han quedado entre 0 y 100
for col in GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_scaled.columns[4:]:
    if (GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_scaled[col].min() != 0):
        print(f"Minimum value of {col} is diferent than 0")
    if (GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_scaled[col].max() < 99) or (GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_scaled[col].max() > 101) :
        print(f"Maximum value of {col} is much diferent than 100")
```

```python
# Me quedo por un lado con las temporadas completas para que sean la base del estudio
GRANDES_5_LIGAS_EUR_FBREF_201617_202223 = GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_scaled[GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_scaled['Temporada']<'2023_2024']
# Me quedo con la temporada en curso para aplicar los resultados del estudio
GRANDES_5_LIGAS_EUR_FBREF_202324_J28 = GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_scaled[GRANDES_5_LIGAS_EUR_FBREF_201617_202324_J28_scaled['Temporada']=='2023_2024']
```

```python
# Me quedo con los campeones de cada liga
CHAMP_GRANDES_5_LIGAS_EUR_FBREF_201617_202223 = GRANDES_5_LIGAS_EUR_FBREF_201617_202223[GRANDES_5_LIGAS_EUR_FBREF_201617_202223['Rk']==1]
```
